```python
#ข้อ 1


import sys


class Graph():

    def __init__(self, vertices):
        self.V = vertices
        self.graph = [[0 for column in range(vertices)]
                      for row in range(vertices)]

    def printSolution(self, dist, minpath):
        ch = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
        print("Vertex \t\tDistance from Source \tMinimum Path")
        for node in range(self.V):
            print(ch[node], "\t\t\t", dist[node], "\t\t", minpath[node])

    def minDistance(self, dist, sptSet):

        min = sys.maxsize

        for v in range(self.V):
            if dist[v] < min and sptSet[v] == False:
                min = dist[v]
                min_index = v

        return min_index

    def dijkstra(self, src):

        dist = [sys.maxsize] * self.V
        dist[src] = 0
        sptSet = [False] * self.V
        minpath = [""] * self.V
        ch = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
        for cout in range(self.V):

            u = self.minDistance(dist, sptSet)
            sptSet[u] = True

            for v in range(self.V):
                if self.graph[u][v] > 0 and sptSet[v] == False and dist[v] > dist[u] + self.graph[u][v]:
                    dist[v] = dist[u] + self.graph[u][v]
                    minpath[v] = minpath[u] + " -> " + str(ch[u])

        self.printSolution(dist, minpath)

g = Graph(7)
g.graph = [
    [0, 3, 0, 6, 1, 0, 0],  # a
    [3, 0, 6, 0, 5, 0, 0],  # b
    [0, 6, 0, 0, 3, 10, 0],  # c
    [6, 0, 0, 0, 6, 0, 7],  # d
    [1, 5, 3, 6, 0, 0, 4],  # e
    [0, 0, 10, 0, 0, 0, 0],  # f
    [0, 0, 0, 7, 4, 0, 0],  # g
]

g.dijkstra(0)
```

```python
#4.4

from binarytree import bst
root = bst(height = 3)
print('Perfect BST of given height : \n',
      root)

class findRes:
    def __init__(self,pre):
        self.left = 0
        self.right = 0
        self.pre = pre
        
    def leftSum(self, round=0):
        if round < len(self.pre):
            if self.pre[round] == None:
                return self.leftSum(round+1)
            elif round == 0:
                return self.leftSum(round+1)
            elif self.pre[round] < self.pre[0]:
                self.left = self.left + self.pre[round]
                return self.leftSum(round+1)
            else:
                self.leftSum(round+1)
        return 0
        
    def rightSum(self,round=0):
        if round < len(self.pre):
            if self.pre[round] == None:
                return self.rightSum(round+1)
            elif round == 0:
                return self.rightSum(round+1)
            elif self.pre[round] > self.pre[0]:
                self.right = self.right + self.pre[round]
                return self.rightSum(round+1)
            else:
                self.rightSum(round+1)
        return 0
        
    
pre = root.values
print(pre)

res = findRes(pre)
res.leftSum()
res.rightSum()
print("ผลรวมด้านซ้าย :" , res.left)
print("ผลรวมด้านขวา :" , res.right)
```
